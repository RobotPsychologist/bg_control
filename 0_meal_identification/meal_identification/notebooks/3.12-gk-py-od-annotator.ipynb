```python
import pandas as pd
import os

def load_interim_data(interim_dir):
    """
    Load all CSV files from the interim directory into a list of pandas DataFrames.
    
    Parameters:
    interim_dir (str): Path to the interim data directory.
    
    Returns:
    list: A list of pandas DataFrames.
    """
    dataframes = []
    for filename in os.listdir(interim_dir):
        if filename.endswith('.csv'):
            filepath = os.path.join(interim_dir, filename)
            df = pd.read_csv(filepath)
            dataframes.append(df)
    return dataframes

# Define the path to the interim data directory
interim_dir = '../data/interim'

# Load the data
dataframes = load_interim_data(interim_dir)

# Display the first few rows of each DataFrame
for i, df in enumerate(dataframes):
    print(f"DataFrame {i}:")
    display(df.head(2))
```

```python
import pandas as pd
import numpy as np

# Convert msg_type using label encoding 
msg_type_mapping = {
    "ANNOUNCE_MEAL": 0,
    "DOSE_INSULIN": 1,
    "DOSE_BASAL_INSULIN": 2,
    "BGL_FP_READING": 3,
    "INTERVENTION_SNACK": 4,
    np.nan: -100  # Handle NaNs
}

for df in dataframes:
    # Convert to timestamp in milliseconds
    # Convert date columns to Unix timestamp in milliseconds
    df['date'] = pd.to_datetime(df['date']).astype('int64') // 10**6
    df['day_start_shift'] = pd.to_datetime(df['day_start_shift']).astype('int64') // 10**6

    df["msg_type_encoded"] = df["msg_type"].map(msg_type_mapping).fillna(-1).astype(int)

    # Convert affects_fob and affects_iob to binary (0/1)
    df["affects_fob"] = df["affects_fob"].fillna(False).astype(int)
    df["affects_iob"] = df["affects_iob"].fillna(False).astype(int)

    # Replace NaN values with 0 
    df['bgl'] = df['bgl'].fillna(0).astype(float)
    df["dose_units"] = df["dose_units"].fillna(0).astype(int)
    df["food_g"] = df["food_g"].fillna(0).astype(int)
    df["food_glycemic_index"] = df["food_glycemic_index"].fillna(0).astype(int)
    df["food_g_keep"] = df["food_g_keep"].fillna(0).astype(int)
    
    # Replace the original 'msg_type' column with the encoded values
    df['msg_type'] = df['msg_type_encoded']
    
    # Drop the temporary 'msg_type_encoded' column 
    df.drop("msg_type_encoded", axis=1, inplace=True)

    df.infer_objects()

    display(df.head(3))
```

```python
from pyod.models.knn import KNN
from sktime.annotation.adapters import PyODAnnotator

# Initialize a list to store each PyODAnnotator model
list_of_models = []
anomalies_dict = {}

for i, df in enumerate(dataframes):
    if not df.empty:
        # Initialize a PyODAnnotator with KNN model and lower contamination
        model = PyODAnnotator(KNN(contamination=0.2))  # Adjust contamination as needed

        # Fit the model using a broader set of features
        model.fit(df[['bgl', 'affects_fob', 'affects_iob', 'dose_units', 'food_g', 'food_glycemic_index', 'food_g_keep']])
        
        # Store the fitted model
        list_of_models.append(model)

        # Use the model to detect anomalies and store the results
        anomalies = model.predict(df[['bgl', 'affects_fob', 'affects_iob', 'dose_units', 'food_g', 'food_glycemic_index', 'food_g_keep']])
        anomalies_dict[f"df_{i}"] = anomalies

        print(f"Anomalies detected in DataFrame {i}:", anomalies.sum())  # Summary of anomalies
```

```python
# Check for a sample of anomalies in the DataFrame
for i, df in enumerate(dataframes):
    anomalies = pd.Series(anomalies_dict[f"df_{i}"], index=df.index[:len(anomalies_dict[f"df_{i}"])])
    anomalies = anomalies.reindex(df.index, fill_value=0)  # Ensure alignment
    
    # Print sample positions and values of anomalies for inspection
    anomaly_positions = anomalies[anomalies == 1].index
    print(f"DataFrame {i} - Anomaly count: {len(anomaly_positions)}")
    print(df.loc[anomaly_positions, ['bgl']].head(10))  # Display first 10 anomalies for inspection
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Loop through each DataFrame and plot bgl values with anomalies highlighted
for i, df in enumerate(dataframes):
    # Extract bgl values
    bgl_values = df['bgl']
    
    # Align anomalies with the DataFrame index
    anomalies = pd.Series(anomalies_dict[f"df_{i}"], index=df.index[:len(anomalies_dict[f"df_{i}"])])
    anomalies = anomalies.reindex(df.index, fill_value=0)  # Ensure anomalies match df's index length

    # Separate normal and anomalous points for plotting
    normal_bgl = bgl_values[anomalies == 0]
    anomalous_bgl = bgl_values[anomalies == 1]
    
    # Create a plot
    plt.figure(figsize=(10, 6))
    
    # Plot all BGL values
    plt.plot(df.index, bgl_values, label='bgl', color='blue', linewidth=0.5)
    
    # Plot anomalies as a separate scatter plot
    plt.scatter(anomalous_bgl.index, anomalous_bgl, color='red', label='Anomaly', s=40)
    
    # Plot settings
    plt.title(f'BGL Values with Anomalies Highlighted - DataFrame {i}')
    plt.xlabel('Record Index')
    plt.ylabel('BGL')
    plt.legend()
    plt.show()
```
